```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# data=pd.read_csv('3_SAT_cooling.csv')
data=pd.read_csv('3_SAT_cooling_20.csv',names=['iteration','n_bits','beta','clauses','overlap'])
data['density']=data.apply(lambda x: x.clauses/x.n_bits, axis=1)
```

```python
df=data.groupby(['n_bits','beta','density'])['overlap'].mean().reset_index()
df['std']=data.groupby(['n_bits','beta','density'])['overlap'].std().reset_index()['overlap']
```

```python
plt.figure(figsize=(10,7))
ax = plt.subplot(111)
for n in [1, 2, 3, 4, 5]:
    plt.plot(df[df['beta']==n]['density'],df[df['beta']==n]['overlap'], label=r'$\beta$=%d'%(n,))
    plt.fill_between(df[df['beta']==n]['density'], df[df['beta']==n]['overlap']-df[df['beta']==n]['std'], 
                    df[df['beta']==n]['overlap']+df[df['beta']==n]['std'])

leg = plt.legend(bbox_to_anchor=(1, 1), ncol=1, shadow=True, fancybox=True)

# leg.get_frame().set_alpha(0.5)
# plt.title('Overlap between thermal states and ground states 12 bits')
plt.xlabel('Density')
ax.axvline(4.27, color='k', linestyle='--')
# ax.axhline(0.5, color='k', linestyle='--')
ax.set_xticks(np.delete(np.arange(0,10,2),2))
ax.set_xlim([0,10])
ax.set_xticks(list(ax.get_xticks()) + [4.27])
plt.savefig('3SAT_15.pdf')

plt.show()
```

```python
plt.figure(figsize=(10,7))
ax = plt.subplot(111)
for n in [1, 2, 3, 4, 5]:
#     plt.scatter(df[df['beta']==n]['density'],df[df['beta']==n]['overlap'] , label=r'$\beta$=%d'%(n,))
    plt.errorbar(df[df['beta']==n]['density'],df[df['beta']==n]['overlap'] , 
                 yerr=df[df['beta']==n]['std'], label=r'$\beta$=%d'%(n,),)
leg = plt.legend(bbox_to_anchor=(1, 1), ncol=1, shadow=True, fancybox=True)

# leg.get_frame().set_alpha(0.5)
plt.title('3SAT cooling for 20 bits')
plt.xlabel('Density')

plt.show()
```

# QAOA

```python
data=pd.read_csv('3_SAT_qaoa_10_025.csv', names=['iteration','p','clauses'])
data['density']=data.apply(lambda x: x.clauses/10, axis=1)
```

```python
df=data.groupby(['density'])['p'].mean().reset_index()
df['std']=data.groupby(['density'])['p'].std().reset_index()['p']
```
