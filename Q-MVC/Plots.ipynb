```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# data=pd.read_csv('3_SAT_cooling.csv')
data=pd.read_csv('MVC_cooling.csv',names=['iteration','n_bits','graph_num','beta','n_0','overlap'])
data['density']=data.apply(lambda x: x.n_0/x.n_bits, axis=1)
```

```python
for_hist=data[(data['beta']==3)&(data['n_bits']==12)]
sns.distplot(for_hist['density'],hist=True )
```

```python
for_hist=data[(data['beta']==3)&(data['n_bits']==15)]
sns.distplot(for_hist['density'],hist=True )
```

```python
for_hist=data[(data['beta']==3)&(data['n_bits']==20)]
sns.distplot(for_hist['density'],hist=True )
```

```python
df=data.groupby(['n_bits','beta','density'])['overlap'].mean().reset_index()
df['std']=data.groupby(['n_bits','beta','density'])['overlap'].std().reset_index()['overlap']
```

```python
plt.figure(figsize=(10,7))
ax = plt.subplot(111)
for n in range(1,6,1):
#     plt.scatter(df[df['beta']==n]['density'],df[df['beta']==n]['overlap'] , label=r'$\beta$=%d'%(n,))
    plt.errorbar(df[(df['beta']==n)&(df['n_bits']==12)]['density'],df[(df['beta']==n)&(df['n_bits']==12)]['overlap'], yerr=df[(df['beta']==n)&(df['n_bits']==12)]['std'],
                label=r'$\beta$=%d'%(n,))
leg = plt.legend(bbox_to_anchor=(1, 1), ncol=1, shadow=True, fancybox=True)

# leg.get_frame().set_alpha(0.5)
plt.title('MVC cooling 12 bits')
plt.xlabel('x_c')

plt.show()
```
